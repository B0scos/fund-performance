```python
import pandas as pd
```

```python
path  = r'..\data\processed\cleaned.parquet'
```

```python
df = pd.read_parquet(path)
```

```python
df = df.sort_values('report_date').reset_index(drop=True)
```

```python
df.head(), df.shape
```

```python
groupy_by_cnpj = df.groupby('fund_cnpj')
```

```python
df['return'] = groupy_by_cnpj['quota_value'].pct_change()
df['total_value_by_net_asset_value'] = df['total_value'] / df['net_asset_value']
```
